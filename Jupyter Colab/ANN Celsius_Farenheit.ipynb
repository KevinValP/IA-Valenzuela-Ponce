```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40, -25, -10, 0, 10, 25, 35, 50], dtype=float)
farenheit = np.array([-40, -13, 14, 32, 50, 77, 95, 122], dtype=float)
```

```python
layer_1 = tf.keras.layers.Dense(units=3, input_shape=[1])
layer_2 = tf.keras.layers.Dense(units=3)
output = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([layer_1, layer_2, output])
```

```python
modelo.compile(
    optimizer = tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
#Entrenar el modelo
history = modelo.fit(celsius, farenheit, epochs=500)
```

```python
#Predicciones contra a mano
print("Predicción")
predictions = modelo.predict([28])
print (predictions)
print ("vs")
print ("A mano")
print(28*1.8 + 32)
```
