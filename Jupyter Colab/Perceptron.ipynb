```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
f = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1), metrics=['mean_squared_error'])
```

```python
history = model.fit(c, f, epochs=500, verbose=False)
print("Se entrenó el modelo")
```

```python
history.params
```

```python
plt.xlabel('Numero de Epocas')
plt.ylabel("Magnitud de la pérdida (loss)")
plt.plot(history.history['loss'])
plt.show()
```

```python
print("Predicción")
print(model.predict([100.0]))
print("vs")
print("A mano")
a_mano=(100*1.8)+32
print(a_mano)
```
